```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.linear_model import Ridge, LinearRegression, Lasso
```

```python
df = pd.read_csv('energydata_complete.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
# Check for null values
df.isna().sum()
```

```python
# Question 12

x = df['T2'].values.reshape(-1,1)
y = df['T6'].values.reshape(-1,1)

x_train,x_test,y_train,y_test= train_test_split(x, y, test_size=0.3, random_state=42)

# linear regression training on x and y
lin_reg=LinearRegression()
lin_reg.fit(x_train,y_train)
predicts =lin_reg.predict(x_test)

# r_square 
from sklearn.metrics import r2_score
r_squared=r2_score(y_test, predicts)
round(r_squared, 2)
```

```python
# dropping date and lights column from the dataset

df.drop(['date', 'lights'], axis = 1, inplace = True)
```

```python
# Normalise the dataset using MinMaxScaler class

scaler = MinMaxScaler()

normalised_df = pd.DataFrame(scaler.fit_transform(df), columns=df.columns)
target_df = normalised_df['Appliances']
features_df = normalised_df.drop('Appliances', axis=1)
```

```python
# Splitting the dataset(70-30) in to training and testing datasets

x_train, x_test, y_train, y_test = train_test_split(features_df, target_df, test_size=0.3, random_state=42)
linear_model = LinearRegression()

# fit the model to the training dataset
linear_model.fit(x_train, y_train)

# obtain predictions
predicted_values = linear_model.predict(x_test)
```

```python
# Question 13

# Mean Square Error(MAE)
mean_a_err = mean_absolute_error(y_test, predicted_values)
round(mean_a_err, 2)
```

```python
# Question 14

# Residual Sum Squares(RSS) 
rss = np.sum(np.square(y_test - predicted_values))
round(rss, 2)
```

```python
# Question 15

# Root mean Squared Error(RMSE)
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

```python
# Question 16

#Co-efficient of determination(R_squared)
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2)
```

```python
# Question 20

# Lasso training
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)

predict_lasso = lasso_reg.predict(x_test)
#RMSE after lasso 
rmse_lasso = np.sqrt(mean_squared_error(y_test, predict_lasso))
round(rmse_lasso, 3)
```

```python
# train ridge model

ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)

#predict ridge
predicted_ridge_values = ridge_reg.predict(x_test)
```

```python
# Question 19
# Let's compare the effects of regularisation

def get_weights_df(model, feat, col_name):
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ['Features', col_name]
    weights_df[col_name].round(3)
    return weights_df

linear_model_weights = get_weights_df(linear_model, x_train,'Linear_Model_Weight')
ridge_weights_df = get_weights_df(ridge_reg, x_train, 'Ridge Weight')
lasso_weights_df = get_weights_df(lasso_reg, x_train, 'Lasso_weight')

final_weights = pd.merge(linear_model_weights, ridge_weights_df, on='Features')
final_weights = pd.merge(final_weights, lasso_weights_df, on='Features')
final_weights
```

From the results above, only 4 features are with non-zero weights

```python
# Question 17

# Query lowest and highest weights
lowest_and_highest = get_weights_df(linear_model,x_train,'Linear model').sort_values(by=['Linear model'],ascending=True)
lowest_and_highest
```

Hence RH2 and RH_1 are features with lowest and highest weights

```python
# Question 18

# RMSE after ridge regression test
rmse_ridge = np.sqrt(mean_squared_error(y_test, predicted_ridge_values))
round(rmse_ridge, 3)
```
